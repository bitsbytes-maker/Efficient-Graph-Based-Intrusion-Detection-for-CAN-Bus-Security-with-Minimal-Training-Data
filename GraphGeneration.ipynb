```python
import matplotlib.pyplot as plt
import networkx as nx
def showAttackFreeGraphFor(graph):
    plt.figure(figsize=(20, 20))
    pos = nx.spring_layout(graph, seed=42)  # `seed` ensures reproducibility
    nx.draw(
        graph, 
        pos, 
        node_size=2000, 
        with_labels=True, 
        width=0.5, 
        font_size=8,
        node_color='green',
        
    )
    plt.savefig("graphAttackfree.png", bbox_inches="tight")
    plt.show()
    
  
# Visualizing the attacked Graph  
def showAttackedGraphFor(attackGraph):

    # List of nodes to highlight
    ndlist = ['000']  

    plt.figure(figsize=(20, 20))  # Adjust figure size for better visualization

    # Use spring_layout or another layout algorithm
    pos = nx.spring_layout(attackGraph, seed=42)  # `seed` ensures reproducibility

    # Draw the entire graph
    nx.draw(
        attackGraph, 
        pos, 
        node_size=1000,  # Adjust node size
        with_labels=True, 
        width=0.5,  # Adjust edge width
        font_size=10, 
        node_color='green',  # Default color for nodes
        edge_color='gray',  # Edge color for better contrast
        alpha=0.9
    )

    # Highlight specific nodes in red
    nx.draw_networkx_nodes(
        attackGraph, 
        pos, 
        node_size=1000,  # Keep consistent node size
        nodelist=ndlist, 
        node_color='red'
    )

    plt.savefig("graphAttacked.png", bbox_inches="tight")
    plt.show()    
```
